Criando um data frame e analisando as vendas de 4 lojas diferentes no mesmo período de tempo, com gráficos separados

```python
import pandas as pd
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df = pd.DataFrame(vendas_2022, index=lojas)
df
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2,2,figsize=(15,10))
fig.suptitle('Vendas 2022', fontsize=20)
ymin = 0
ymax = 450
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)
  ax.set_xlabel('Meses')
  ax.set_ylabel('Vendas')

axs[0,0].plot(df.loc['A'], lw=3, color='blue')
axs[0,0].set_title('Loja A', loc='left', fontsize=18)

axs[0,1].plot(df.loc['B'], lw=3, color='green')
axs[0,1].set_title('Loja B',loc='left',fontsize=18)

axs[1,0].plot(df.loc['C'], lw=3, color='purple')
axs[1,0].set_title('Loja C',loc='left',fontsize=18)

axs[1,1].plot(df.loc['D'], lw=3, color='yellow')
axs[1,1].set_title('Loja D',loc='left',fontsize=18)
plt.show()
```

- Importando banco de dados com informações referêntes a imigração de diversos países para o canadá;
- Separando as informações do Brasil, Argentina, Peru e Colômbia entre os anos de 1980 a 2013;
- Criando um data frame com essas informações que foram separadas;
- Criando um gráfico para análisar tendências, considerando país, título, rótulos nos eixos, cores apropriadas,  tamanho das fontes e espessura das linhas.

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df.set_index('País', inplace=True)
```

```python
df
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes':brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
argentina = df.loc['Argentina', anos]
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes':argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
```

```python
peru = df.loc['Peru', anos]
peru_dict = {'ano': peru.index.tolist(), 'imigrantes':peru.values.tolist()}
dados_peru = pd.DataFrame(peru_dict)
```

```python
colombia = df.loc['Colômbia', anos]
colombia_dict = {'ano': colombia.index.tolist(), 'imigrantes': colombia.values.tolist()}
dados_colombia = pd.DataFrame(colombia_dict)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,8))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw = 2)
plt.plot(dados_argentina['ano'], dados_argentina['imigrantes'], lw = 2)
plt.plot(dados_peru['ano'], dados_peru['imigrantes'], lw = 2)
plt.plot(dados_colombia['ano'], dados_colombia['imigrantes'], lw = 2)
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'], fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('Ano', fontsize = 16)
plt.ylabel('Número de imigrantes', fontsize = 16)
plt.title('Número de imigrantes para o Canadá entre 1980 a 2013', fontsize = 20, loc='left', pad=30)
plt.legend(['Brasil', 'Argentina', 'Peru', 'Colômbia'], fontsize = 14)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(color='lightgrey')

```
